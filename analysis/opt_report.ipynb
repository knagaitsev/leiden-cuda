```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import json
from pathlib import Path

curr_path = Path()

with open(curr_path / "../results/opt_report.json", "r") as f:
    runtime_results = json.load(f)

df = pd.DataFrame(runtime_results)

f, ax = plt.subplots(1, figsize=(4, 3))

labels = df["label"]
values = df["runtime"]

# plt.title("cuGraph Leiden varied max_iter")
plt.xlabel("Optimizations")
plt.ylabel('Runtime (s)')

plt.bar(labels, values, color="#ffa600", linewidth=1, edgecolor='black')
# plt.plot(max_iter, values)

plt.xticks(rotation=20, ha='right')
plt.grid(axis='y')
plt.gca().set_axisbelow(True)

plt.subplots_adjust(left=0.2, bottom=0.3)

plt.yscale('log')

figs_dir = Path.resolve(curr_path / "../figs")

os.makedirs(figs_dir, exist_ok=True)

plt.savefig(figs_dir / f"opt_report.pdf")
```
